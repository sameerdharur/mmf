```python
for i in range(xy.shape[0]):
    vec1 = np.array(json.loads(xy['reasoning_importance_vector_1'][i]))
    vec2 = np.array(json.loads(xy['reasoning_importance_vector_2'][i]))
    vec1 = vec1/np.max(vec1)
    vec2 = vec2/np.max(vec2)
    vec3 = np.concatenate((vec1, vec2))
    xy['reasoning_importance_vector_1'][i] = vec1
    xy['reasoning_importance_vector_2'][i] = vec2
    xy['reasoning_importance_vector_3'][i] = vec3
    vec4 = np.array(json.loads(xy['sub_question_importance_vector_1'][i]))
    vec5 = np.array(json.loads(xy['sub_question_importance_vector_2'][i]))
    vec4 = vec4/np.max(vec4)
    vec5 = vec5/np.max(vec5)
    vec6 = np.concatenate((vec4, vec5))
    xy['sub_question_importance_vector_1'][i] = vec4
    xy['sub_question_importance_vector_2'][i] = vec5
    xy['sub_question_importance_vector_3'][i] = vec6
    xy['similarity_1'][i] = np.dot(vec1, vec4)/(np.linalg.norm(vec1) * np.linalg.norm(vec4))
    xy['similarity_2'][i] = np.dot(vec2, vec5)/(np.linalg.norm(vec2) * np.linalg.norm(vec5))
    xy['similarity_3'][i] = np.dot(vec3, vec6)/(np.linalg.norm(vec3) * np.linalg.norm(vec6))
```
